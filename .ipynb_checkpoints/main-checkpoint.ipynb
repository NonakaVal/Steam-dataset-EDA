### Requirements

```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import os
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
from dateutil import parser
from collections import defaultdict
OPENAI_API_KEY = os.getenv(key='OPENAI_API_KEY')
```

```python
def _chat(df_path):
    """Langchain Function to dataframe AI chat
    """
    openai_api_key = OPENAI_API_KEY
    df = df_path
    llm = ChatOpenAI(model = "gpt-3.5-turbo", openai_api_key=openai_api_key)
    agent_prefix = "O df importado é\n"

    agent = create_pandas_dataframe_agent(
        llm,
        df,
        prefix=agent_prefix,
        verbose  = True,
        agent_type = AgentType.OPENAI_FUNCTIONS,
    )

    while True:
        prompt = input("Digite o prompt ou 'sair\n'")
        if prompt.lower() == 'sair':
            print("Finalizado...")
            break
        response = agent.invoke(prompt)
```

#### Variables
- app_id - This is a unique identifier for each game.
- title - This indicates the name of the game.
- release_date - This is the date when the game was released.
- genres - These are the categories the game falls under.
- categories - These can be extra characteristics about the game (like if it's single-player, multiplayer, has full controller support etc.).
- developer - This is the name of the company or individual who developed the game.
- publisher - This is the name of the company or individual who published the game.
- original_price - This represents the price of the game before any discounts.
- discount_percentage - This portrays the percentage discount offered on the game.
- discounted_price - This is the price of the game after applying the discount.
- dlc_available - This indicates whether the game has downloadable content (DLC) available or not.
- age_rating - This suggests the recommended age for players of the game.
- content_descriptor - These are notes or labels describing the type of content present in the game which prompted the assigned age rating.
- about_description - This provides a brief description regarding what the game is about.
- win_support, mac_support, linux_support - These indicate whether the game is compatible with Windows, Mac, and Linux platforms respectively.
- awards - This represents the number of awards the game has received.
- overall_review - This displays the overall rating of the game based on user reviews.
- overall_review_% - This portrays the ratio of positive to negative reviews in percentage.
- overall_review_count - This shows the total number of reviews received by the game.
- recent_review - This presents the rating of the game based on recent user reviews.
- recent_review_% - This shows the ratio of recent positive to negative reviews in percentage.
- recent_review_count - This refers to the number of recent reviews the game received01

### Cleaning

```python
# new df without null values = new_df
to_drop = ['original_price', 'discount_percentage', 'content_descriptor', 
                          'recent_review', 'recent_review_%', 'recent_review_count']
new_df = df.drop(columns=to_drop).dropna()
```

```python
# convert dates to datetime and overall_review_count to int
def convert_date(df):
    try:
        return pd.to_datetime(parser.parse(df))
    except:
        return pd.NaT 
new_df['release_date'] = new_df['release_date'].apply(convert_date).dropna()

new_df['overall_review_count'] = new_df['overall_review_count'].astype(int)
```

```python
# discounted_price to int
new_df['discounted_price'] = new_df['discounted_price'].replace('Free', '0')
new_df['discounted_price'] = new_df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Função para extrair valores distintos da variável = contagem_distinta(df, coluna)
def contagem_distinta(df, coluna):
    """
    Calcula a contagem distinta de valores em uma coluna específica de um DataFrame.

    Parâmetros:
        df (pandas.DataFrame): O DataFrame contendo os dados.
        coluna (str): O nome da coluna que contém os valores a serem contados.

    Retorna:
        dict: Um dicionário contendo os valores únicos como chaves e suas contagens como valores.
    """
    # Array de valores
    valores_array = df[coluna].values
    
    # Dicionário para armazenar a contagem de valores
    contagem_valores = defaultdict(int)
    
    # Iterar sobre cada linha e contar os valores
    for item in valores_array:
        if isinstance(item, str):
            for valor in item.split(", "):
                contagem_valores[valor] += 1
        else:
            contagem_valores[item] += 1
                
    return dict(contagem_valores)
```

```python
# contagem de generos distintos.
distinct_genres_count = contagem_distinta(new_df, 'genres')
sorted_genres_count = sorted(distinct_genres_count.items(), key=lambda x: x[1], reverse=True)
for genre, count in sorted_genres_count:
    print(f"{genre}: {count}")
```

```python
genres_count_df = pd.DataFrame(sorted_genres_count, columns = ['genre', 'count'])
```

```python
# Definindo o estilo do gráfico
sns.set_style("whitegrid")

# Criando uma paleta de cores personalizada
custom_palette = sns.color_palette("viridis", len(genres_count_df))

# Criando o gráfico de barras horizontal
plt.figure(figsize=(15, 10))
sns.barplot(data=genres_count_df, x='count', y='genre', palette=custom_palette)

# Adicionando título e rótulos aos eixos
plt.title('Distribuição de Jogos por Gênero', fontsize=16)
plt.xlabel('Quantidade', fontsize=14)
plt.ylabel('Gênero', fontsize=14)

# Ajustando o tamanho da fonte dos ticks nos eixos
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adicionando a contagem dentro de cada barra
for index, value in enumerate(genres_count_df['count']):
    plt.text(value, index, str(value), fontsize=12, va='center', ha='left')

# Removendo a borda do gráfico
sns.despine()

# Exibindo o gráfico
plt.show()
```

```python
# Contagem Categorias Distintas
distinct_categories = contagem_distinta(new_df, 'categories')

# Ordenar os valores do dicionário
sorted_categories_count = sorted(distinct_categories.items(), key=lambda x: x[1], reverse=True)

# Imprimir os resultados ordenados
for genre, count in sorted_categories_count:
    print(f"{genre}: {count}")
```

```python
review_by_date = new_df[['overall_review', 'release_date']]
```

```python
review_map = {
    'Overwhelmingly Positive': 9,
    'Very Positive': 8,
    'Positive': 7,
    'Mostly Positive': 6,
    'Mixed': 5,
    'Mostly Negative': 4,
    'Negative': 3,
    'Very Negative': 2,
    'Overwhelmingly Negative': 1
}
```

```python
# Criando uma cópia independente do DataFrame original
review_by_date_copy = review_by_date.copy()

# Substituindo os valores da coluna 'overall_review' na cópia
review_by_date_copy['overall_review'] = review_by_date_copy['overall_review'].replace(review_map)
```

```python
# Extrair o ano da data de lançamento
review_by_date_copy['year'] = review_by_date_copy['release_date'].dt.year

# Calcular a média das avaliações por ano
mean_by_year = review_by_date_copy.groupby('year')['overall_review'].mean()

# Plotar o gráfico de tendência
plt.figure(figsize=(15, 8))
mean_by_year.plot(marker='o', linestyle='-')
plt.title('Tendência da Avaliação ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Avaliação Média')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='year', y='overall_review', data=review_by_date_copy, palette='muted')
plt.title('Distribuição das Avaliações por Ano')
plt.xlabel('Avaliação')
plt.ylabel('Ano')
plt.grid(True)
plt.show()
```

```python
# Plotando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(review_by_date_copy['release_date'], review_by_date_copy['overall_review'], alpha=0.5)
plt.title('Gráfico de Dispersão entre Overall Review e Release Date')
plt.xlabel('Release Date')
plt.ylabel('Overall Review')
plt.grid(True)
plt.show()
```
